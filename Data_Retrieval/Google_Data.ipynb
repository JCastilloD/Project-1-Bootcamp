```python
#DENUE_df = pd.read_csv("Resources/DENUE_Filtered_complete.csv")
DENUE_df = pd.read_csv("Output/DENUE_Filtered.csv")
display(DENUE_df)
long_and_lat_df = pd.read_csv("Resources/Long_and_Lat_by_municip.csv")
display(long_and_lat_df)
```

```python
Size_dict={'0 a 5 personas':1,
           '6 a 10 personas':2,
           '11 a 30 personas':3,
           "31 a 50 personas":4,
           '51 a 100 personas':5,
           '101 a 250 personas':6,
           '251 y más personas':7}
Size_dict2={'0 a 5 personas':"Micro",
           '6 a 10 personas':"Micro",
           '11 a 30 personas':"Pequeña",
           "31 a 50 personas":"Pequeña",
           '51 a 100 personas':"Mediana",
           '101 a 250 personas':"Mediana",
           '251 y más personas':"Grande"}

DENUE_df["Company Size"]=DENUE_df["Descripcion estrato personal ocupado"].astype(str)
DENUE_df["Company Size (int)"]=DENUE_df["Descripcion estrato personal ocupado"].astype(str)
DENUE_df=DENUE_df.replace({"Company Size (int)": Size_dict})
DENUE_df=DENUE_df.replace({"Company Size": Size_dict2})


DENUE_df
```

```python
DENUE_df = pd.merge(DENUE_df, long_and_lat_df, on="Clave municipio", how="left")
DENUE_df
```

```python
Micro_df = DENUE_df.loc[DENUE_df["Company Size"]=="Micro"]
locations_Micro = Micro_df[["Latitud", "Longitud"]]

Small_df = DENUE_df.loc[DENUE_df["Company Size"]=="Pequeña"]
locations_Small = Small_df[["Latitud", "Longitud"]]

Medium_df = DENUE_df.loc[DENUE_df["Company Size"]=="Mediana"]
locations_Medium = Medium_df[["Latitud", "Longitud"]]

Large_df = DENUE_df.loc[DENUE_df["Company Size"]=="Grande"]
locations_Large = Large_df[["Latitud", "Longitud"]]


rating_Micro = Micro_df["Company Size (int)"]#.astype(float)
rating_Small = Small_df["Company Size (int)"]
rating_Medium = Medium_df["Company Size (int)"]#.astype(float)
rating_Large = Large_df["Company Size (int)"]
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create layers
#symbols = gmaps.symbol_layer(locations, fill_color='red', stroke_color='red')
heat_layer1 = gmaps.heatmap_layer(locations_Micro, weights=rating_Micro, 
                                 dissipating=False, max_intensity=1,
                                 point_radius=0.005,gradient = [(0,0,0,0),"blue","blue"])

heat_layer2 = gmaps.heatmap_layer(locations_Small, weights=rating_Small, 
                                 dissipating=False, max_intensity=1,
                                 point_radius=0.005,gradient = [(0,0,0,0),"yellow","yellow"])

heat_layer3 = gmaps.heatmap_layer(locations_Medium, weights=rating_Medium, 
                                 dissipating=False, max_intensity=1,
                                 point_radius=0.005,gradient = [(0,0,0,0),"green","green"])

heat_layer4 = gmaps.heatmap_layer(locations_Large, weights=rating_Large, 
                                 dissipating=False, max_intensity=1,
                                 point_radius=0.005,gradient = [(0,0,0,0),"red","red"])



# Add layer
fig.add_layer(heat_layer1)
fig.add_layer(heat_layer2)
fig.add_layer(heat_layer3)
fig.add_layer(heat_layer4)
#fig.add_layer(symbols)
# Display figure
fig
```
