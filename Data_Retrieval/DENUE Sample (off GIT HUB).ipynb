```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

%matplotlib inline
from matplotlib import pyplot as plt
import scipy.stats as stats

# Import API key
#from config import inegi_key #inegi token
#from config import coneval_key #coneval
#from config import gkey #google key
```

```python
denue_df = pd.read_csv("DENUE_full_data.csv",encoding="ISO-8859-1")
#denue_df = pd.read_csv("Resources/DENUE_full_data_old.csv",encoding="ISO-8859-1")
denue_df
```

```python
denue_df_sample = denue_df.sample(frac=.12)
denue_df_sample
```

```python
countmpo = pd.DataFrame(denue_df["Clave municipio"].value_counts())
countmpo["Cvemposam"] = pd.DataFrame(denue_df_sample["Clave municipio"].value_counts())
countmpo["Cvempototpct"] = countmpo["Clave municipio"] / countmpo["Clave municipio"].sum()
countmpo["Cvemposampct"] = countmpo["Cvemposam"] / countmpo["Cvemposam"].sum()

display(countmpo)
```

```python
critical_value = stats.chi2.ppf(q = 0.95, df = 124)
display(critical_value)
stats.chisquare(countmpo["Cvempototpct"], countmpo["Cvemposampct"])
```

```python
#Statistic is below critical value of 150, so we conclude the sample is not different to the overall
```

```python
countestr = pd.DataFrame(denue_df["Descripcion estrato personal ocupado"].value_counts())
countestr["Cveestrsam"] = pd.DataFrame(denue_df_sample["Descripcion estrato personal ocupado"].value_counts())
countestr["Cveestrtotpct"] = countestr["Descripcion estrato personal ocupado"] / countestr["Descripcion estrato personal ocupado"].sum()
countestr["Cveestrsampct"] = countestr["Descripcion estrato personal ocupado"] / countestr["Descripcion estrato personal ocupado"].sum()

display(countestr)
```

```python
critical_value = stats.chi2.ppf(q = 0.95, df = 6)
display(critical_value)
stats.chisquare(countestr["Cveestrtotpct"], countestr["Cveestrsampct"])
```

```python
#Statistic is below critical value of 150, so we conclude the sample is not different to the overall
```
